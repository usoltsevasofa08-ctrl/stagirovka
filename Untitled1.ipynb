```python
# --- Подготовка данных ---
# 1. Преобразуем столбцы с датами в формат datetime
order['order_purchase_timestamp'] = pd.to_datetime(order['order_purchase_timestamp'])
order['order_delivered_customer_date'] = pd.to_datetime(order['order_delivered_customer_date'])

# 2. Фильтруем только доставленные заказы
delivered_orders = order[order['order_status'] == 'delivered'].copy()

# 3. Рассчитываем время доставки в днях
delivered_orders['delivery_time_days'] = (delivered_orders['order_delivered_customer_date'] - 
                                          delivered_orders['order_purchase_timestamp']).dt.days

# 4. Удаляем аномалии (если есть отрицательные значения или слишком большие выбросы)
# Например, оставляем только заказы со временем доставки от 0 до 100 дней
delivered_orders_clean = delivered_orders[(delivered_orders['delivery_time_days'] >= 0) & 
                                          (delivered_orders['delivery_time_days'] <= 100)]

print(f"Всего доставленных заказов: {len(delivered_orders)}")
print(f"После удаления аномалий: {len(delivered_orders_clean)}")

# --- Построение графиков ---
plt.figure(figsize=(14, 6))

# Гистограмма (распределение времени доставки)
plt.subplot(1, 2, 1)
sns.histplot(delivered_orders_clean['delivery_time_days'], bins=30, kde=True, color='#570ca1')
plt.title('Распределение времени доставки заказов', fontsize=14)
plt.xlabel('Время доставки (дни)', fontsize=12)
plt.ylabel('Количество заказов', fontsize=12)
# Добавляем вертикальную линию для среднего значения
mean_delivery = delivered_orders_clean['delivery_time_days'].mean()
plt.axvline(mean_delivery, color='red', linestyle='--', label=f'Среднее: {mean_delivery:.1f} дней')
plt.legend()

plt.tight_layout()
plt.show()
print('Среднее время доставки', int(mean_delivery), 'дней')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

# Загрузка данных
orders_df = pd.read_csv('olist_orders_dataset.csv')
customers_df = pd.read_csv('olist_customers_dataset.csv')

# Объединяем таблицы
merged_df = pd.merge(orders_df, customers_df, on='customer_id', how='inner')

print("=" * 70)
print("ШАГ 1: СОЗДАНИЕ МАТРИЦЫ ПРИЗНАКОВ ПОЛЬЗОВАТЕЛЕЙ")
print("=" * 70)

# Создаем признаки для каждого пользователя
user_features = merged_df.groupby('customer_id').agg({'order_id': 'count',  # количество заказов
}).rename(columns={'order_id': 'total_orders'}).reset_index()

# Добавляем признак - количество уникальных штатов
user_states = merged_df.groupby('customer_id')['customer_state'].nunique().reset_index(name='unique_states')
user_features = pd.merge(user_features, user_states, on='customer_id', how='left')

# Добавляем признак - количество уникальных городов
user_cities = merged_df.groupby('customer_id')['customer_city'].nunique().reset_index(name='unique_cities')
user_features = pd.merge(user_features, user_cities, on='customer_id', how='left')

# Добавляем признак - среднее время между заказами (если несколько заказов)
# Сортируем по дате для каждого пользователя
merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'])
merged_df_sorted = merged_df.sort_values(['customer_id', 'order_purchase_timestamp'])

# Вычисляем разницу между заказами
merged_df_sorted['prev_order_date'] = merged_df_sorted.groupby('customer_id')['order_purchase_timestamp'].shift(1)
merged_df_sorted['days_between_orders'] = (merged_df_sorted['order_purchase_timestamp'] - merged_df_sorted['prev_order_date']).dt.days

# Среднее время между заказами для каждого пользователя
avg_days = merged_df_sorted.groupby('customer_id')['days_between_orders'].mean().reset_index(name='avg_days_between_orders')
user_features = pd.merge(user_features, avg_days, on='customer_id', how='left')

# Заполняем пропуски (для пользователей с 1 заказом ставим 0)
user_features = user_features.fillna(0)

print(f"Количество пользователей: {len(user_features)}")
print(f"Количество признаков: {len(user_features.columns) - 1}")
print("\nПервые 5 строк матрицы признаков:")
print(user_features.head())

# --- Шаг 2: Преобразование в матрицу X ---

print("\n" + "=" * 70)
print("ШАГ 2: ПРЕОБРАЗОВАНИЕ В МАТРИЦУ X И МАСШТАБИРОВАНИЕ")
print("=" * 70)

# Создаем матрицу X (без customer_id)
feature_columns = ['total_orders', 'unique_states', 'unique_cities', 'avg_days_between_orders']
X = user_features[feature_columns].values

print(f"Размер матрицы X: {X.shape} (пользователи × признаки)")

# Масштабирование признаков
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\nСтатистика до масштабирования:")
print(pd.DataFrame(X, columns=feature_columns).describe())

print("\nСтатистика после масштабирования:")
print(pd.DataFrame(X_scaled, columns=feature_columns).describe())

# --- Шаг 3: Вычисление косинусной близости ---

print("\n" + "=" * 70)
print("ШАГ 3: ПОИСК НАИБОЛЕЕ ПОХОЖИХ ПОЛЬЗОВАТЕЛЕЙ")
print("=" * 70)

# Берем выборку для вычислений (если пользователей много)
if len(X_scaled) > 500:
    print(f"Пользователей много ({len(X_scaled)}), беру первые 500 для примера...")
    X_sample = X_scaled[:500]
    user_ids_sample = user_features['customer_id'].iloc[:500].values
else:
    X_sample = X_scaled
    user_ids_sample = user_features['customer_id'].values

# Вычисляем матрицу косинусной близости
similarity_matrix = cosine_similarity(X_sample)

# Заполняем главную диагональ нулями (чтобы не находить самого себя)
np.fill_diagonal(similarity_matrix, 0)

# Находим максимальную близость
max_sim_idx = np.unravel_index(np.argmax(similarity_matrix), similarity_matrix.shape)
user1_idx, user2_idx = max_sim_idx[0], max_sim_idx[1]
max_similarity = similarity_matrix[user1_idx, user2_idx]

user1_id = user_ids_sample[user1_idx]
user2_id = user_ids_sample[user2_idx]

print(f"\n✅ Наиболее похожая пара пользователей:")
print(f"  Пользователь 1: {user1_id}")
print(f"  Пользователь 2: {user2_id}")
print(f"  Косинусная близость: {max_similarity:.4f}")

# Показываем признаки этих пользователей
user1_features = user_features[user_features['customer_id'] == user1_id].iloc[0]
user2_features = user_features[user_features['customer_id'] == user2_id].iloc[0]

print("\n📊 Признаки пользователя 1:")
print(f"  Заказов: {user1_features['total_orders']}")
print(f"  Уникальных штатов: {user1_features['unique_states']}")
print(f"  Уникальных городов: {user1_features['unique_cities']}")
print(f"  Среднее время между заказами: {user1_features['avg_days_between_orders']:.1f} дней")

print("\n📊 Признаки пользователя 2:")
print(f"  Заказов: {user2_features['total_orders']}")
print(f"  Уникальных штатов: {user2_features['unique_states']}")
print(f"  Уникальных городов: {user2_features['unique_cities']}")
print(f"  Среднее время между заказами: {user2_features['avg_days_between_orders']:.1f} дней")

# --- Шаг 4: Матрица корреляций (цветная таблица) ---

print("\n" + "=" * 70)
print("ШАГ 4: МАТРИЦА КОРРЕЛЯЦИЙ МЕЖДУ ПРИЗНАКАМИ")
print("=" * 70)

# Вычисляем корреляцию
correlation_matrix = pd.DataFrame(X_scaled, columns=feature_columns).corr()

print("Матрица корреляций (цветная таблица):")

# Стилизованная цветная таблица
plt.figure(figsize=(8, 6))
sns.heatmap(
    correlation_matrix, 
    annot=True,           # цифры внутри ячеек
    fmt='.3f',            # 3 знака после запятой
    cmap='coolwarm',      # синий-белый-красный
    center=0,             # белый = 0
    square=True, 
    linewidths=0.5,
    vmin=-1, vmax=1       # фиксированная шкала
)
plt.title('Матрица корреляций', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# --- Шаг 5: Дополнительная визуализация ---

print("\n" + "=" * 70)
print("ШАГ 5: ВИЗУАЛИЗАЦИЯ ПОЛЬЗОВАТЕЛЕЙ")
print("=" * 70)

# Выбираем до 50 пользователей для визуализации
n_users = min(50, len(X_sample))

if n_users > 1:
    from sklearn.decomposition import PCA
    
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_sample[:n_users])
    
    plt.figure(figsize=(10, 8))
    
    # Строим scatter plot с цветовой кодировкой по количеству заказов
    orders_colors = X_sample[:n_users, 0]  # первый признак - количество заказов
    
    scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], 
                         c=orders_colors, cmap='viridis', 
                         s=100, alpha=0.7)
    
    plt.colorbar(scatter, label='Количество заказов')
    plt.title('Визуализация пользователей в 2D (PCA)', fontsize=14, fontweight='bold')
    plt.xlabel(f'Первая главная компонента ({pca.explained_variance_ratio_[0]:.1%} дисперсии)')
    plt.ylabel(f'Вторая главная компонента ({pca.explained_variance_ratio_[1]:.1%} дисперсии)')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Находим похожих пользователей для случайного
    random_idx = np.random.randint(0, n_users)
    random_user = user_ids_sample[random_idx]
    similarities = similarity_matrix[random_idx]
    top5_indices = np.argsort(similarities)[-5:][::-1]
    
    print(f"\n🔍 Топ-5 пользователей, похожих на {random_user}:")
    for i, idx in enumerate(top5_indices, 1):
        if idx != random_idx:
            sim_score = similarities[idx]
            similar_user = user_ids_sample[idx]
            print(f"  {i}. {similar_user} (близость: {sim_score:.4f})")

# --- Шаг 6: Интерпретация ---

print("\n" + "=" * 70)
print("ВЫВОДЫ И ИНТЕРПРЕТАЦИЯ")
print("=" * 70)

print(f"""
1. МАТРИЦА ПРИЗНАКОВ:
   - Создана матрица размером {X.shape[0]} × {X.shape[1]}
   - Признаки: количество заказов, уникальные штаты, уникальные города, среднее время между заказами

2. КОСИНУСНАЯ БЛИЗОСТЬ:
   - Максимальная близость: {max_similarity:.4f}
   - Это {'высокий' if max_similarity > 0.7 else 'средний' if max_similarity > 0.4 else 'низкий'} уровень схожести
   - Наиболее похожие пользователи имеют схожее поведение в заказах

3. КОРРЕЛЯЦИИ ПРИЗНАКОВ:
   {correlation_matrix.to_string()}

4. ПРАКТИЧЕСКОЕ ПРИМЕНЕНИЕ:
   - На основе близости можно рекомендовать товары: "пользователи, похожие на вас, также заказывали..."
   - Корреляции помогают понять, какие признаки связаны между собой
   - Это основа для построения recommendation systems
""")
```
